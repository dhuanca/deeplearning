## Conclusion

```python
# Definir datos de resultados
modelos = ['Modelo 1', 'Modelo 2', 'Modelo 3', 'Modelo 4', 'Modelo 5', 'Modelo6','Modelo7']
rmse = [rmse_1,rmse_2, rmse_3, rmse_4, rmse_5, rmse_6, rmse_7]
mape = [mape_1, mape_2, mape_3, mape_4, mape_5,mape_6, mape_7]


# Crear DataFrames
df_configuracion = pd.DataFrame({
    'Modelo': modelos,
    'Configuración': [
        'time_steps = 1',
        'time_steps = 7',
        'time_steps = 30',
        'time_steps = 30 y 3 variables { Close, MA(3), MA(7) }',
        'time_steps = 30 y 4 Variables { Close, Volumen, MA(3), MA(7) }',
        'time_steps = 30 , pred_steps=5 y 4 Variables { Close, Volumen, MA(3), MA(7) }',
        'time_steps = 30 , pred_steps=5, 4 Variables { Close, Volumen, MA(3), MA(7) } y 64 neuronas '
    ]
})


df_resultados = pd.DataFrame({
    'Modelo': modelos,
    'RMSE': rmse,
    'MAPE': mape
})

# Ordenar por MAPE
df_final = df_configuracion.merge(df_resultados, on='Modelo').sort_values('MAPE')

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
df_final.set_index('Modelo')[['RMSE', 'MAPE']].plot(kind='bar', rot=0)
plt.title('Gráfico de Barras de RMSE y MAPE')
plt.xlabel('Modelo')
plt.ylabel('Valor')
plt.legend(loc='upper left')
plt.show()

```
